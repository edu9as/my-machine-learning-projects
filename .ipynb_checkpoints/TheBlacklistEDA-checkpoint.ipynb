## Loading libraries

Also, to represent the data as a dataframe I am using **Pandas**, and to visualize the data I am writing some basic plots with **matplotlib.pyplot**.
````python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import re
import time

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None) 
````

## Defining functions

To facilitate the extraction of the data, I am defining multiple functions:

- **get_tables** takes a URL from Wikipedia, gets the webpage under that URL and looks for the tables where the episodes data is stored (hint: these data are within <table> tags with attribute class = "wikiepisodetable").
- **set_season** assigns the corresponding season number to each episode. 
- **get_pd_dfs** convert the HTML tables to pandas dataframes, and labels these dataframes with the corresponding season (**set_season**). Also, sets the row number to the global episode number. Returns a list of pandas DataFrames.
- **df_from_list** takes a list of pandas DataFrames and joins them in an unique DataFrame. Also cleans the data in the DataFrame:
    - Removes the quotation marks in the Title column.
    - Removes references from the numbers (between brackets) in US Viewers column.
    - Converts the Blacklist IDs to only numbers.
- **flatten_list** takes a nested list and converts it to a flat list. Useful to complement the next function.
- **add_info_from_imbd** takes the dataframe generated with Wikipedia information and adds some more information from IMDb, such as IMDb rating and a brief episode description.

Because it is not funny to suffer a spoiler, I have taken into account this situation and an option is included in **dfs_from_list** and **add_info_from_imdb** so that you can avoid seeing information you don't want to see. If you are currently watching season 6 (indeed, this is my case), you can set ```avoid_spoiler = 6``` in these two functions so that the information of season 6 and later is not displayed.

````python
def get_tables(url):
    page = requests.get(url).text
    soup = bs(page, "html.parser")
    tables = soup.findAll("table", {"class": "wikiepisodetable"})
    return tables

def set_season(df_list):
    for i in range(len(df_list)):
        df_list[i].insert(0, "Season", i + 1)
        
    return df_list

def get_pd_dfs(table_list):
    dfs = [pd.read_html(str(table))[0] for table in table_list]
    dfs = set_season(dfs)
    [df.set_index('No.overall', inplace = True) for df in dfs]
    return dfs

def df_from_list(df_list, avoid_spoiler = False):
    if avoid_spoiler:
        end = avoid_spoiler - 1
    else:
        end = 7
    df = pd.concat(df_list[:end])
    df.reindex(list(range(len(df))))

    df["Original air date"] = pd.to_datetime(df["Original air date"])
    blacklist_id = np.zeros(len(df))
    blacklist = list(df["Blacklistguide"].values)
    titles = list(df["Title"])
    viewers = list(df['US viewers(millions)'])
    for i in range(len(df)):
        try:
            titles[i] = re.search('"(.+)"', titles[i]).groups()[0]
            viewers[i] = re.search('(.+)\[', viewers[i]).groups()[0]
            blacklist_id[i] = int(re.search("(\d+)", blacklist[i]).groups()[0])
        except:
            continue
    df["Blacklistguide"] = blacklist_id
    df["Title"] = titles
    df['US viewers(millions)'] = viewers
    df = df.astype({"Blacklistguide": int,
                    "US viewers(millions)": float})
        
    return df

def flatten_list(nested_list):
    flat_list = []
    for subl in nested_list:
        for item in subl:
            flat_list.append(item)
    return flat_list

def add_info_from_imdb(df, avoid_spoiler = False):
    imdb_url = "https://www.imdb.com/title/tt2741602/episodes?season={}"
    pages = []

    if avoid_spoiler:
        end = avoid_spoiler

    else:
        end = 8

    for i in range(1, end):
        pages.append(requests.get(imdb_url.format(i)).text)
        time.sleep(3)
    
    all_eps = [bs(page, "html.parser").findAll("div", class_="list_item") for page in pages]
    all_imdb = [[float(episode.find("span", "ipl-rating-star__rating").text) for episode in season] for season in all_eps]
    all_descr = [[episode.find("div", itemprop="description").text.strip() for episode in season] for season in all_eps]

    df["Description"] = flatten_list(all_descr)
    df["IMDb rating"] = flatten_list(all_imdb)

    return df
````

## Getting the data!

Thanks to the functions defined above, getting the data is a very easy task. By entering the URL of the Blacklist episodes Wikipedia webpage, the HTML tables are extracted, then they are converted to Pandas dataframes and merged into an only dataframe, and this dataframe is completed with information from IMDb.

````python
url_en = "https://en.wikipedia.org/wiki/List_of_The_Blacklist_episodes"

df = add_info_from_imdb(df_from_list(get_pd_dfs(get_tables(url_en))))
````

### Numerical summary
Let's see a numerical summary of this dataset:

````python
df.describe().T
````

There are five numerical columns:
- Season: discrete variable which holds values between 1 and 7 (because season 8 is not complete, I have omitted it).
- No. inseason: discrete variable which contains values between 1 and 23 (all seasons have an episode number 1 although not all of them have an episode number 23).
- Blacklistguide: the IDs of the Blacklist that appear in each episode. In a few episodes there is more than 1 Blacklist ID (alway contiguous), but I have kept only the lowest ID. In some other episodes there is not any Blacklist ID, so I have labeled these as 0. The rest of IDs range between 1 and 184.
- US viewers (in millions): the millions of people in the USA that viewed each episode. Because it is a continuous variable, the mean and standard deviation here make sense: the average number of viewers in the US is 6.66 million, with an standard deviation of 2.86 millions. The minimum number of viewers was 3.36 millions, while the maximum number was over 25 millions (this number seems an outlier, we will explore it later).
- IMDb rating: continuous variable with a mean over 8 (I think it is a pretty good metric though I'm not an IMDb expert) and standard deviation of only 0.5. The episode with a lower IMDb rating got a 6, while that with the highest got a 9.4.

## Questions to answer with an image

Finally, I will try to answer some possible questions for these data with an only (or two in some cases) plot.

### How many episodes are there in every season? How many episodes with a given number are there?


````python
q1 = "How many episodes are there per season?"
q2 = "How many episodes #20 are there?"

kws = [[0,"Season", 7, "#7777FF", q1, "Season #"],
       [1,"No. inseason", 23, "#3333AA", q2, "Season episode #"]]

fig, ax = plt.subplots(1,2, figsize = (20,5))

for kw in kws:
    bins = np.linspace(0.5, kw[2] + 0.5, kw[2] + 1)
    df.hist(kw[1], bins = bins, color = kw[3], ax = ax[kw[0]], edgecolor = "white", grid = False)

    if kw[0] == 0:
        yticks = [10,19,20,22,23]
        xticks = list(range(1,8))
    else:
        yticks = list(range(1,8))
        xticks = list(range(1,24))

    plt.xticks(xticks)
    plt.yticks(yticks)
    ax[kw[0]].set_xlabel(kw[5])
    ax[kw[0]].set_ylabel("Number of episodes")
    ax[kw[0]].set_title(kw[4])

plt.show()
````

### In which season appears each Blacklist ID?

````python
fig, ax = plt.subplots(1,1,figsize = (20,5))
for thing in zip(list(range(1,8)), ["red", "orange", "green", "blue", "purple", "black", "brown"]):
    df[df["Season"] == thing[0]].hist("Blacklistguide", color = thing[1], 
                                      bins = np.linspace(-0.5, 184.5, 185), ax = ax, 
                                      legend = True, grid = False)
plt.legend(labels= ["Season " + str(i) for i in range(1,8)])
plt.xticks(np.linspace(0,180, 37))
plt.xlim(1,185)
plt.yticks([0,1,2,3])
plt.title("In which season appears each Blacklist ID?")
plt.ylabel("Blacklist ID appearences")
plt.xlabel("Blacklist ID")
plt.show()
````

### Which are the names of the most important Blacklisters (lowest Blacklist ID) and in which season they are dealt with? Arbitrary threshold set at 30

````python
def get_labels_for_blacklist_ids(df):
    titles = list(df["Title"])
    blids = list(df["Blacklistguide"])

    labs = {str(blids[i]): titles[i] for i in range(len(titles)) 
            if blids[i] > 0 and "Conclusion" not in titles[i]}

    return labs

labs = get_labels_for_blacklist_ids(df)

fig, ax = plt.subplots(1,1,figsize = (20,5), dpi = 300)

threshold = 30

for thing in zip(list(range(1,8)), ["red", "orange", "green", "blue", "purple", "black", "brown"]):
    df[df["Season"] == thing[0]].hist("Blacklistguide", color = thing[1], 
                                      bins = np.linspace(-0.5, 184.5, 185), ax = ax, 
                                      legend = True, grid = False)

for k, v in labs.items():
    if int(k) > threshold:
        continue
    y = len(df[df["Blacklistguide"] == int(k)]) + 0.25
    plt.text(int(k), y, v, rotation = 90, size = 12)

plt.legend(bbox_to_anchor=(1,1), loc="upper left",
           labels= ["Season " + str(i) for i in range(1,8)])

plt.xlim(0.5,threshold + 0.5)
plt.ylim(0,9)
plt.xticks(list(range(1,threshold + 1)))
plt.yticks([0,1,2,3])

plt.xlabel("Blacklist ID")
plt.ylabel("Blacklist ID appearences")
plt.title("Which are the names of the most important Blacklisters (lowest Blacklist IDs) "
          "and which season they are dealt with? (first {} Blacklisters)".format(threshold))

plt.show()
````

### Is the Blacklist ID related to the IMDb ratings?

````python
plt.figure(figsize = (10,5))
plt.plot(df["Blacklistguide"], df["IMDb rating"], "o")
plt.xlabel("Blacklist ID")
plt.ylabel("IMDb rating")
plt.title("Is the Blacklist ID related to the IMDb ratings?")
plt.xlim(1, 186)
plt.ylim(6.9,9.5)
plt.show()
````

````python
plt.figure(figsize = (10,5))

season_color = zip(list(range(1,8)), 
                   ["red", "orange", "green", "blue", "purple", "black", "brown"])

for season, color in season_color:
    subset = df[df["Season"] == season]
    x = list(range(1,185))
    z = np.polyfit(subset["Blacklistguide"], subset["IMDb rating"], 1)
    p = np.poly1d(z)
    plt.plot(x,p(x),"--", color = color)
    plt.plot(subset["Blacklistguide"], subset["IMDb rating"], "o", color = color)

plt.xlabel("Blacklist ID")
plt.ylabel("IMDb rating")
plt.title("Is the Blacklist ID related to the IMDb ratings? With trend lines per season")
plt.legend(bbox_to_anchor=(1,1), loc="upper left",
           labels= [["Season " + str(i//2), ""][i%2 == 1] for i in range(1,16)], numpoints = 1)
plt.xlim(1, 186)
plt.ylim(6.9,9.5)
plt.show()
````
